```python
#constants
R0 = 37
T0 = 22.1
alpha = 4.5E-3
min_deviation = 48.363
sensor0 = np.round(262.67+25/180, 3)
```

```python
#function to find the maximum index corresponding to the highest intensity
def maximumIndex(array):
    return (i for i in range(len(array)) if array[i] == max(array))
```

```python
#calcualte the temperature here
T_220V = None
angles_220V, intensities_220V = np.loadtxt("./Trial1.txt", unpack = True, delimiter = None)
angles_220V_real = angles_220V - (sensor0)
refractive_indices_220V = np.sqrt(((2/np.sqrt(3))*np.sin(angles_220V_real)+0.5)**2+0.75)
print(refractive_indices_220V)
plt.scatter(angles_220V_real, intensities_220V)

print("\n")
print(refractive_indices_220V[list(maximumIndex(intensities_220V))[0]])
```
