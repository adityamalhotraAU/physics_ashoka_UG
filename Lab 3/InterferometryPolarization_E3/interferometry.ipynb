```python
import numpy as np
import matplotlib.pyplot as plt
from fits_library import *
```

$$\text{Part A: Wavelength of laser used}$$

```python
#least count of screw guage 0.5/50 = 0.02 mm
#count the number of dark or bright fringes that collapse into the central circle for equal intervals of movement. Plot the two together

#centre of the fringe is dark

C = 0.0241

delta_d = np.array([2, 6, 10, 12, 14, 18, 21, 23, 25, 32, 36, 39, 41, 43, 47, 52, 57, 60, 62, 65]) * 0.01
fringes_collapsed = np.array([2, 5, 7, 10, 12, 15, 17, 20, 22, 25, 27, 30, 32, 35, 37, 40, 42, 45, 47, 50])

m_wavelength, c_wavelength, m_wavelength_err, c_wavelength_err = linear_fit(delta_d, fringes_collapsed)

fit_wavelength = m_wavelength * delta_d + c_wavelength

print((2 * C)/m_wavelength * 10E5, len(delta_d), m_wavelength, c_wavelength)

plt.scatter(delta_d, fringes_collapsed)
plt.plot(delta_d, fit_wavelength, '--', color = 'firebrick')
plt.errorbar(delta_d, fringes_collapsed, xerr = 0.01, yerr = 1, ls = 'none', color = 'k', capsize = 1.5)
plt.xlabel(r"$\Delta d$ (mm)")
plt.ylabel("No. of fringes collapsed")
plt.savefig("laser_wavelength.png")
plt.show()
```
