```python
import numpy as np
import matplotlib.pyplot as plt
```

$\text{Bode Plot for high-pass filter}$

```python
log_freq, gain = np.loadtxt("./high_pass_bode.txt", unpack = True)
```

```python
#Error for high pass
vin1, vout1 = np.loadtxt("./vin_vout_high_pass.txt", unpack = True)
err_highpass = 20*np.sqrt((0.16/vin1**2)+(16/vout1**2))/(np.log(10))
plt.scatter(log_freq, err_highpass)
plt.xlabel(r"Frequency in log scale")
plt.ylabel(r"Relative error in measuring gain")
gain_err1 = np.abs(err_highpass)
for i in range(0,len(err_highpass)):
    print(np.round(gain[i],2), "&", np.round(err_highpass[i], 2), "\\\\", "\hline")
```

```python
plt.scatter(log_freq, gain, color = 'firebrick')
plt.plot(log_freq, gain, '--', color = 'blue')
plt.axvline(3.08,color = 'tomato', label = r"$\omega = \omega_c$")
plt.xlabel(r"Frequency (log scale)")
plt.ylabel(r"Gain (dB)")
plt.errorbar(log_freq, gain, yerr = gain_err1, ecolor= 'black', capsize = 3, ls = "none")
plt.legend()
plt.show()
```

$\text{Bode plot for LCR ``tank" configuration}$

```python
lcr_freq_tank, gain_tank = np.loadtxt("./lcr_tank.txt", unpack = True)
```

```python
#error in lcr tank
vin4, vout4 = np.loadtxt("./vin_vout_lcr_tank.txt", unpack = True)
err_lcr_tank = 20*np.sqrt((0.16/vin4**2)+(16/vout4**2))/(np.log(10))
plt.scatter(lcr_freq_tank, err_lcr_tank)
plt.xlabel(r"Frequency in log scale")
plt.ylabel(r"Relative error in measuring gain")
gain_err4 = np.abs(err_lcr_tank*gain_tank)
for i in range(0,len(err_lcr_tank)):
    print(np.round(gain_tank[i],2), "&", np.round(err_lcr_tank[i]*100, 2), "\\\\", "\hline")
```

```python
plt.scatter(lcr_freq_tank, gain_tank, color = 'firebrick')
plt.plot(lcr_freq_tank, gain_tank, '--' ,color = 'blue')
plt.axvline(3.48, color = 'tomato', label = r"$\omega = \omega_0$")
plt.xlabel(r"Frequency (log scale)")
plt.ylabel(r"Gain (dB)")
plt.errorbar(lcr_freq_tank, gain_tank, yerr = gain_err4, ecolor = 'black', capsize = 3, ls = 'none')
plt.legend()
plt.show()
```
