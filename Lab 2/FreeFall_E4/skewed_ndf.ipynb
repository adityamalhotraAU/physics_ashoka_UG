```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, skewnorm
```

```python
array = np.array([9.68568398,  9.54956955,  9.81671821,  9.87625834,  9.56954877,  9.61155935,
  9.7212955,   9.69388203,  9.53087602,  9.7532484,  10.5113455,  10.5940058,
 10.25945087, 10.14442312, 10.50484111,9.62745323, 10.27344135, 10.2050894,
 10.01635515, 10.6694763,  10.31731484,  9.92177322, 10.48097409,  9.85369223,])
skewness = skew(array)
print(skewness)
```

```python
numargs = skewnorm .numargs  
a, b = -0.347, 10.007
rv = skewnorm (a, b)  
print(rv)
```

```python
quantile = np.arange (9.2, 10.4, 0.1)  
  
# Random Variates  
R = skewnorm.rvs(a, b)  
print ("Random Variates : \n", R)  
  
# PDF  
R = skewnorm.pdf(a, b, quantile)  
print ("\nProbability Distribution : \n", R)  
```

```python
distribution = np.linspace(0, np.minimum(rv.dist.b, 3))  
print("Distribution : \n", distribution)  
     
plot = plt.plot(distribution, rv.pdf(distribution))  
```
