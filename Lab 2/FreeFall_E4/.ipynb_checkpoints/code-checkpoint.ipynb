```python
tb1, yb1, vyb1 = np.loadtxt("./part b data/0.15g.txt", unpack = True, skiprows=2)
plt.scatter(tb1, yb1)
plt.show()
vtfitb1_m, vtfitb1_c, vtfitb1_m_error, vtfitb1_c_error = linear_fit(tb1[-15:], yb1[-15:])
print(np.abs(vtfitb1_m))
vel_tanh1 = vtfitb1_m*np.tanh((9.8*tb1)/vtfitb1_m)
plt.plot(tb1, vel_tanh1)
plt.scatter(tb1,-vyb1)
plt.show()
corr_matrix1 = np.corrcoef(-vyb1, vel_tanh1)
corr1 = corr_matrix1[0,1]
R_sq1 = corr1**2
print(R_sq1)
```
