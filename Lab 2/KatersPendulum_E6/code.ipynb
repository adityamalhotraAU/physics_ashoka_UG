```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.dpi"] = 150
```

```python
l, T = np.loadtxt("./partA_data.txt", unpack = True)
T = T/10
```

```python
plt.scatter(l ,T)
plt.axhline(1.68, ls = '--', color = 'firebrick', label = 'Equal time periods')
plt.axvline(-45, ls = '--', color = 'tomato', label = 'Points on rod')
plt.axvline(-40, ls = '--', color = 'tomato')
plt.axvline(40, ls = '--', color = 'tomato')
plt.axvline(30, ls = '--', color = 'tomato')
plt.legend(loc = 'upper right')
plt.errorbar(l, T, xerr = 0.1, yerr = 0.01, capsize = 2.5, ls = 'none', color = 'black')
plt.show()
```
