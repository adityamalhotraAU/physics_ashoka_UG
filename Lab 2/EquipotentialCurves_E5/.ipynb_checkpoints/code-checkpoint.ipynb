$$V(y) = \biggl(\frac{V_0}{d}\biggl)y$$

```python
voltage1 = np.array([0.56, 0.87, 1.34, 1.85, 2.16])
distance1 = (np.array([4-2.5, 4-1.5,4- 0.1, 4+1.4, 4+2.6]))
m, c, m_err_rel, c_err_rel = linear_fit(distance1, voltage1)
fiteq1 = m*distance1 + c
plt.scatter(distance1, voltage1)
plt.plot(distance1, fiteq1, '--', color = 'firebrick')
plt.errorbar(distance1, voltage1, xerr = 0.3, yerr=0.01, capsize = 1.5, ecolor='black', ls='none')
plt.xlabel('Distance from Grounded Electrode (cm)')
plt.ylabel('Potential (V)')
print(m, m_err_rel*100, m*m_err_rel*100, c)
plt.show()
corr_matrix1 = np.corrcoef(distance1, voltage1)
corr1 = corr_matrix1[0,1]
R_sq1 = corr1**2
print(R_sq1)
```

$$V(r) = \frac{(V_1-V_0)}{\ln({\frac{r_1}{r_0}})}\ln\biggl({\frac{r}{r_0}}\biggl) + V_0$$

```python
radius1 = np.mean(r1)
radius2 = np.mean(r2)
radius3 = np.mean(r3)
radius4 = np.mean(r4)
radius5 = np.mean(r5)
radius_arr = np.array([radius1, radius2, radius3, radius4, radius5])/100
log_radius_arr = np.log(radius_arr/R0)
voltage_arr = np.array([v1, v2, v3, v4, v5])
```

```python
dr = (0.3*np.sqrt(2))/100
dV0 = 0.01
dV = np.sqrt(((-3.0206267288172826/radius_arr)*dr)**2+(0.13804954335734426)**2)
print(dV)
delr = dr/radius_arr
print(delr)
```

```python
plt.scatter((log_radius_arr), voltage_arr)
m2, c2, m2_err_rel, c2_err_rel = linear_fit(log_radius_arr, voltage_arr)
fiteq2 = m2*log_radius_arr + c2
plt.plot(log_radius_arr, fiteq2, '--', color = 'firebrick')
plt.errorbar(log_radius_arr, voltage_arr, xerr = delr, yerr = dV, ecolor = 'k', capsize = 1.5, ls = 'none')
plt.xlabel(r"Log of radius of equipotential curve ($\log{m}$)")
plt.ylabel("Potential (V)")
plt.show()
print(m2, c2, c2*c2_err_rel)
corr_matrix2 = np.corrcoef(log_radius_arr, voltage_arr)
corr2 = corr_matrix2[0,1]
R_sq2 = corr2**2
print(R_sq2)
```
