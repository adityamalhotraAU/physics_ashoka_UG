```python
from numpy import *
import matplotlib.pyplot as plt
```

```python
Vin = array([3.61,3.6,3.6,3.6,3.6,3.6,3.6,3.6,3.59,3.6,3.6,3.6,3.6,3.6,3.6,3.6,3.6,3.6,3.57,3.58,3.57,3.53])
Vout = array([34.2,33.9,33.5,32.8,32.4,31.2,29.5,28.5,27.1,25.2,23.8,21.9,19.4,17.9,16,13.8,12.2,10.8,9.83,8.57,7.83,7.16])
omega = 2*pi*array([500,800,1100,1500,1750,2250,3000,3500,4000,4800,5500,6500,8000,9000,10500,13000,15000,17500,20000,23000,26000,30000])
gain = 20*log10(Vout/Vin)
log_omega = log10(omega)
```

```python
delVin = 5
delVout = 50
delomega = 10**(-3)

delGain = sqrt((delVin/Vin)**2+(delVout/Vout)**2)
```

```python
plt.scatter(log_omega, gain)
plt.xlabel(r'Log of Angular Frequency ($\log({Hz})$)')
plt.ylabel(r'Gain (dB)')
plt.axvline(4.27, color = 'firebrick', ls = '--', label = r'$\omega = \omega_c$')
plt.errorbar(log_omega, gain, xerr = abs(log(delomega)/omega), yerr = delGain, ls = 'none', color = 'black', capsize = 2.5)
plt.title("Bode Plot for Integrator RC Circuit")
plt.legend()
plt.show()
```
